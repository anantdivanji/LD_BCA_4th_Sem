# Simple Linear Regression

## Step 1 : Loading the standard libraries"

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

## Step 2 : Load the data

```python
data = pd.read_csv("Salary_data.csv")
data.head()
```

```python
data.shape
```

## Step 3 : Data Cleaning, Data Preprocessing

```python
sns.regplot(x = data['YearsExperience'], y = data['Salary'])
```

## Step 4 : Seperate X and y

```python
X = data.drop('Salary', axis = 1)
y = data['Salary']
```

## Step 5 : Split the X and y into train test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

## Step 6 : Apply LinearRegression on X_train and y_train

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr
```

```python
lr.fit(X_train, y_train)
```

## Step 7: Perform predictions on X_test

```python
y_pred = lr.predict(X_test)
y_pred
```
