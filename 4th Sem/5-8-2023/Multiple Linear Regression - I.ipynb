# Multiple Linear Regression

## Step 1 : Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

## Step 2 : Load the data

```python
data = pd.read_csv('Advertising.csv')
data.head()
```

```python
data.shape
```

## Step 3 : Data Cleaning & Data Preprocessing

```python
sns.regplot(x = data['TV'], y = data['Sales'])
```

```python
sns.regplot(x = data['Radio'], y = data['Sales'])
```

```python
sns.regplot(x = data['Newspaper'], y = data['Sales'])
```

```python
sns.heatmap(data.corr(), annot = True)
```

## Dropping Newspaper from the data as it has the least relation to Sales       

```python
data = data.drop('Newspaper', axis= 1)
data.head()
```

## Step 4: Seperate X and y

```python
X = data.drop('Sales', axis = 1)
y = data['Sales']
```

## Step 5 : Split X and y into train and test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

## Step 6 : Apply Linear Regression on X_train and y_train

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr
```

```python
lr.fit(X_train, y_train)
```

## Step 7: Perform predictions on X_test

```python
y_pred = lr.predict(X_test)
y_pred
```
