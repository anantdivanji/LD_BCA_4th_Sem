## Fill the missing value with some constant (Constant Imputation)

```python
data = pd.read_csv('Data.csv')
data
```

```python
data.isnull().sum()
```

```python
data['Age'] = data['Age'].fillna(40)
data
```

```python
data['Salary'] = data['Salary'].fillna(50000)
data
```

## Mean Imputation

```python
data = pd.read_csv('Data.csv')
data
```

```python
data['Age'].mean()
```

```python
data['Age'] = data['Age'].fillna(38.77)
data
```

```python
data['Salary'] = data['Salary'].fillna(data['Salary'].mean())
data
```

## Median Imputation

```python
data = pd.read_csv('Data.csv')
data
```

```python
data['Age'] = data['Age'].fillna(data['Age'].median())
data
```

```python
data['Salary'] = data['Salary'].fillna(data['Salary'].median())
data
```

## Backward Fill Imputation

```python
data = pd.read_csv('Data.csv')
data
```

```python
data['Age'] = data['Age'].fillna(method = 'bfill')
data
```

## Forward fill Imputation

```python
data['Salary'] = data['Salary'].fillna(method = 'ffill')
data
```

## Mode Imputation

```python
data = pd.read_csv('Data.csv')
data
```

```python
data['Age'] = data['Age'].fillna(data['Age'].mode())
data
```

```python
data['Salary'] = data['Salary'].fillna(data['Salary'].mode())
data
```

## Missing value impuatation using sklearn library

```python
from sklearn.impute import SimpleImputer
sim_imp = SimpleImputer()
sim_imp
```

```python
data = pd.read_csv('Data.csv')
data
```

```python
data['Age'] = sim_imp.fit_transform(data[['Age']])
data
```

```python
data['Age'].mean()
```

## Median Imputation using SimpleImputer

```python
from sklearn.impute import SimpleImputer
sim_imp = SimpleImputer(strategy = 'median')
sim_imp
```

```python
data['Salary'] = sim_imp.fit_transform(data[['Salary']])
data
```

## Imputation using mode from sklearn imputer

```python
from sklearn.impute import SimpleImputer
sim_imp = SimpleImputer(strategy = 'most_frequent')
sim_imp
```
